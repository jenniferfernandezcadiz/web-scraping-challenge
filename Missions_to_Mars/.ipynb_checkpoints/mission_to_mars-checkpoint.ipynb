# Mars Hemispheres 

1. Visit the USGS Astrogeology site here to obtain high resolution images for each of Mar's hemispheres.


2. You will need to click each of the links to the hemispheres in order to find the image url to the full resolution image.


3. Save both the image url string for the full resolution hemisphere image, and the Hemisphere title containing the hemisphere name. Use a Python dictionary to store the data using the keys img_url and title.


4. Append the dictionary with the image url string and the hemisphere title to a list. This list will contain one dictionary for each hemisphere.

```python
# Cerberus Hempisphere
url_cerberus='https://astrogeology.usgs.gov/search/map/Mars/Viking/cerberus_enhanced'
```

```python
cer_result=r.get(url_cerberus)
cerberus_result=cer_result.text
soup_cerberus=BeautifulSoup(cerberus_result, "html.parser")
```

```python
soup_cerberus
```

```python
cerberus_image=soup_cerberus.find_all('div', class_='wide-image-wrapper')
print(cerberus_image)
```

```python
for c in cerberus_image:
    c_image=c.find('li')
    full_image=c_image.find('a')['href']
#     print(full_image)

c_title=soup_cerberus.find('h2', class_='title').text
# print(c_title)

cerberus_hemp={"Title": c_title, "url": full_image}
cerberus_hemp
```

```python
# Schiaparelli Hempisphere

url_schi='https://astrogeology.usgs.gov/search/map/Mars/Viking/schiaparelli_enhanced'
```

```python
schi_result=r.get(url_schi)
schi_result1=schi_result.text
soup_schi=BeautifulSoup(schi_result1, 'html.parser')
```

```python
soup_schi
```

```python
schi_image=soup_schi.find_all('div', class_='wide-image-wrapper')
print(schi_image)
```

```python
for s in schi_image:
    s_image=s.find('li')
    full_image=s_image.find('a')['href']
#     print(full_image)

s_title=soup_schi.find('h2', class_='title').text
# print(s_title)

schi_hemp={"Title": s_title, "url": full_image}
schi_hemp
```
